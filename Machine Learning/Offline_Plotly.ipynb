```python
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
```

```python
trace0 = go.Scatter(
    x = [1,2,3,4],
    y = [10,15,13,17]
)

trace1 = go.Scatter(
     x = [1,2,3,4],
    y = [10,14,19,12]
)

data = [trace0,trace1]

iplot(data, filename = 'basic-line',image_width=100,image_height=100)
```

```python
trace = go.Scatter(
    x = [1,2,3,4],
    y = [10,15,13,17],
    mode = 'markers'
)
data = [trace]
iplot(data,filename = 'basic-line')
```
